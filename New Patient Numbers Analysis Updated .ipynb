The two most important columns in this dataset are "Average number minus ideal" and "Proportion of actual to target." These two columns are what I consider the performance metrics for each clinic, that is, they provide us a picture as to how well each clinic is doing in terms of bringing in new patients. Let's take a look at the first metric "Average number minus ideal."

```python
df3[['Location', 'Average number of new patients per month', 'ideal',
       'Average number minus ideal']]
```

The columns "Average number minus ideal" is straightforwardly the difference between the average number of new patients per month and the ideal monthly count for each clinic. It is important to note here that the ideal numbers are updated in accordance to Dr. Alexis and therefore the data here is also going to be different than the data from the first report. In general, the ideal numbers were raised, which means that most clinics did not perform as strongly as we had thought at the time of the first report. Nevertheless, as we will see later on incorporating the data from 2022, some clinics actually performed better! Let's take a look at the second performance metric "Proportion of actual to target."

```python
df3[['Location', 'Number of New Patients per 100000 residents',
       'Target Number per 100000 residents', 'Proportion of actual to target']]
```

The column "Proportion of actual to target" is the percentage of number of new patients per 100000 residents to the target number per 100000 residents. How are each of these columns created? The number of new patients per 100000 is created by multiplying the columns "Average number of new patients per month" by "Number of Months of Data", dividing by the population of the zip code in which the clinic is located in, and then multiplying by 100000. This column signifies the number of new patients each clinic would bring in if each zip code had exactly 100000 residents. The purpose of this column is to compare how well each clinic does in relation to each other. By standardizing each zip code's population, we can compare each clinics interrelational performance on an equal playing field. Below is a dataset of the locations with the column "Number of New Patients per 100000 residents" sorted from greatest to least and a corresponding bar chart.

```python
df3.sort_values(by = 'Number of New Patients per 100000 residents', ascending = False)[['Location', 'Number of New Patients per 100000 residents']]
```

```python
labels = df3.sort_values(by = 'Number of New Patients per 100000 residents', ascending = False)['Location']
values = df3.sort_values(by = 'Number of New Patients per 100000 residents', ascending = False)['Number of New Patients per 100000 residents']
plt.figure(figsize=(12,6))
plt.bar(labels, values, color ='maroon', 
        width = 0.4)
plt.axhline(y = df3['Number of New Patients per 100000 residents'].mean().astype('int'), label = 'Average')
plt.xticks(rotation=30, ha='center')
plt.xlabel('Location')
plt.ylabel('Count')
plt.legend(bbox_to_anchor = (0.90, 1), loc = 'upper center')
plt.title('Number of New Patients per 100000 residents')
plt.show()
```

By far, SE PDX brings in the most new patients per 100000 residents, followed by Clackamas and Beaverton. The first six clinics are above the average of 503 new patients per 100000, while the rest are below the average. This provides us with one way of prioritizing clinics by placing priority on those which bring in the most new patients per 100000 residents. This is not a metric meant to motivate us to *improve* certain clinics' performances, but rather to *continue* maintaining the highest-performing clinics. It would be false to assume that this metric would be an adequate way to measure performance since not all clinics have the same ideal count.

Related to the column "Number of New Patients per 100000 residents" is the column "Target Number per 100000 residents," which is calculated by multiplying the ideal by the number of months of data, dividing by the population, and then multiplying by 100000. This is the number of new patients per 100000 each clinic would bring in if it were to meet its ideal count every month. We can combine both of these columns and display them in the following bar chart:

```python
plt.rcParams["figure.figsize"] = (15, 6)
df3.sort_values(by = 'Number of New Patients per 100000 residents', ascending = False).plot(x="Location", y=["Number of New Patients per 100000 residents", "Target Number per 100000 residents"], kind="bar")
plt.xticks(rotation=30, ha='center')
```

From the bar chart above, we can get a rough idea as to how well these clinics are performing in regards to attracting new patients. The closer the blue bar is to the orange bar, the better. Moreover, those clinics which have blue bars higher than their orange bars are over-performing, namely the clinics Clackamas and Gresham. This leads us to the second performance metric which I mentioned earlier, which is "Proportion of actual to target." This is simply the column "Number of New Patients per 100000 residents" divided by the column "Target Number per 100000 residents" and multiplied by 100. Below is a dataset and bar chart of this performance metric:

```python
df3.sort_values(by = 'Proportion of actual to target')[['Location', 'Proportion of actual to target']]
```

```python
labels = df3.sort_values(by = 'Proportion of actual to target')['Location']
values = df3.sort_values(by = 'Proportion of actual to target')['Proportion of actual to target']
plt.figure(figsize=(15,6))
plt.bar(labels, values, width = .4, color = 'maroon')
plt.xticks(rotation=30, ha='center')
plt.axhline(y = 100, color = 'g', linestyle = '-')
plt.axhline(y = 85, color = 'b', linestyle = '-')
plt.xlabel('Location')
plt.ylabel('Percentage')
plt.title('Proportion of actual to target')
plt.show()
```

Here we can a clear picture of the performance of each clinic with respect to this particular metric. For the year 2023, Gresham and Clackamas greatly performed, Beaverton and NE PDX met their exact quotas, while Keizer, SE PDX, and Tigard were very close to meeting their ideal counts. The horizontal green line corresponds to 100%, while the horizontal blue line corresponds to 85%. This is to partition each clinic in terms of priority of improvement: those above green line are doing very well, those between the blue and green line are doing well but could still improve their numbers slightly, and those below the blue line take the most priority.

A couple of things to note here: first, why bother with examining data from just 2023 and data from both 2022 and 2023? The reason why I decided to separate this data is because many businesses reorganize their budgets come the new year. Given that, we can analyze the new patient data given a reorganizing of the budget for just the year 2023, and then shift our attention to data from the past two years. When we ponder what might or might not influence a particular clinic's ability to attract new patients, two things come to mind: competition and reputation. Simplying the situation drastically, one could assume that the less competition a clinic has within the same zip code, the more new patients it should attract. Furthermore, reputation may also potentially be a factor; the more reputable a clinic, the more it can stand out above the competition, and therefore attract new patients. Related to these two aspects are the columns "Number of competing chiropractic clinics in zip code per 100000 residents", "Number of months in operation", "Average Google review score", and "Number of Google reviews."

Let's first take a look at the column "Number of competing chiropractic clinics in zip code per 100000 residents".

```python
both_years_df[['Location', 'Number of competing chiropractic clinics in zip code per 100000 residents']]
```

The question we can seek to answer is: what relationship is there between a clinic's performance and the number of competitors it faces? We will take a look at the correlation between the number of competing chiropractic clinics and the performance metrics "Average Number minus ideal" and "Proportion of actual to target."

```python
both_years_df['Average Number minus ideal'].corr(both_years_df['Number of competing chiropractic clinics in zip code per 100000 residents'])
```

```python
both_years_df['Proportion of actual to target'].corr(both_years_df['Number of competing chiropractic clinics in zip code per 100000 residents'])
```

![correlation](4cf283_2b6b0ceabfa94bd0a7bb80a4d9f99084~mv2.png)

The table above shows the strength of the relationship between two relationships based upon their correlation coefficient. We can see from cells 22 and 23 that our performance metrics "Average Number minus ideal" and "Proportion of actual to target" have correlations of about 0.3, which means that there is only a weak relationship between these metrics and the number of competing chiropractic clinics within the same zip code. What does this mean precisely? One implication is that we can proceed with our strategy of investing more of the budget to the worse-performing clinics without paying much heed to the local competition. What may be more of a factor is not so much the *number* of competition a clinic has, but the *quality* of the competition and the reputation of our clinics as well. This is where the column "Average Google review score" and "Number of Google reviews" comes into play.

```python
both_years_df[['Location', 'Average Google review score', 'Number of Google reviews']]
```

```python
both_years_df['Average Google review score'].mean()
```

There are a few noteworthy observations: first, the average Google review score for all clinics is 4.97, and the minimum score is 4.9. Besides how impressive this is, what this is surely going to mean is that there is not going to be a strong correlation between a clinic's performance and its reputation (at least according to its Google review scores), since virtually all clinics have a perfect score. Let's verify our assumption below:

```python
both_years_df['Average Number minus ideal'].corr(both_years_df['Average Google review score'])
```

```python
both_years_df['Proportion of actual to target'].corr(both_years_df['Average Google review score'])
```

These metrics have only a moderate correlation with the average Google review score of each clinic. Keep in mind that the number of Google reviews matters as well. Some clinics have very few reviews, and responder bias may also be a factor as well (perhaps only those who really enjoyed their experience will leave behind a review). What does this suggest? A clinic's ability to attract new patients does not necessarily depend on the local competition or its average Google review score. We must look to other factors if we are to establish explicit influences on a clinic's ability to attract new patients.
