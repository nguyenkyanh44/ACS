```python
def clean_df(df):
    columns_to_drop = []
    for i in df.columns:
        if all(df[i].isnull()) == True:
            columns_to_drop.append(i)
    df.drop(columns_to_drop, inplace = True, axis = 1)
    df.at[0, 'Unnamed: 1'] = 'RB Frequency'
    df.at[0, 'Unnamed: 12'] = 'M Frequency'
    df.at[0, 'Unnamed: 0'] = 'Referral Breakdown'
    df.at[0, 'Unnamed: 11'] = 'Marketer'
    df.columns = df.loc[0]
    df.drop(0, inplace = True)
    df.reset_index(drop = True, inplace = True)
    df.fillna(0, inplace = True)
    df = df.astype({'RB Frequency' : 'int', 'M Frequency' : 'int'})
```
